# Synthesis

## Sine Wave

In the instance below, we first import the necessary libraries. Numpy lets us create waves of numbers (including binary). sounddevice can take number arrays and generate audio waves from them. time is a default library for time.

```python
import numpy as np
import sounddevice as sd
import time
```

Now create a variable for the samples per second, as well as the time in seconds we want our audio wave to be. Then of course specify the decimal value of your frequency. (or frequencies if you're modulating or adding effects)

```python
sample_rate = 44100
duration = 3.0
frequency = 440.0
```

Now we create an array of 0s that is the length of our total samples. (sample rate times seconds)

```python
x = np.linspace(0, duration * 2 * np.pi, int(duration * sample_rate))
```

Now to generate the sine wave we get the sine of the wave frequency multiplied by the array.

```python
sinewave_data = np.sin(frequency * x)
```

This is the full audio signal with no resistance, so we simply multiply the wave data times some decimal between 0 and 1. 

```python
sinewave_data = sinewave_data * 0.4
```

Now I'll use a data visualation library to slice 500 samples from our wave and plot them. 

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')

plot_data = sinewave_data[:500]

fig, ax = plt.subplots()
ax.plot(plot_data, linewidth=3)
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

## Frequency Modulation 

#### Variables
+ sample_rate = integer, samples per second
+ duration = expressed as a floating decimal value (not integer)
+ frequencyA = The frequency of your "carrier"
+ fmA = the frequency of your modulat0r

```python
sample_rate = 44100
duration = 4.0
frequencyA = 440.0
fmA = 92.3333   
```

We need a "ramp", or logspace that lets us change the carrier phase and modulate the frequency simulateneously as a function of time. 

```python
ramp = np.logspace(0, 1, int(duration * sample_rate))
```

```python
x = np.linspace(0, duration * 2 * np.pi, int(duration * sample_rate))
wave_data = np.sin(frequencyA * x * ramp * np.sin(fmA * x))
wave_data *= 0.2
```

```python
plot_data[:750]

fig, ax = plt.subplots()

ax.plot(plot_data, linewidth=3)
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.show()
```
